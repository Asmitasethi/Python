###Creating a new column named total_bedroom_size. If the size is 10 or less,its quoted as small. If the size is greater than 11 and less than 1000,its quoted as medium; otherwise large.

```python
def total_bedroom_size(x):
  if x<=10:
    return"small"
  elif x>10 and x<=1000 :
    return"medium"
  else:
    return"large"
```

```python
df1["total_bedroom_size"]=df1["total_bedrooms"].apply(total_bedroom_size)
df1
```
